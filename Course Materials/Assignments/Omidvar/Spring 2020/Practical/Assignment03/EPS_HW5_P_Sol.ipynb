```python
import numpy as np
import matplotlib.pyplot as plt
from math import *

def f(x):
    return exp(-log(x)**2/2)/(sqrt(2*pi)*x)

x = np.linspace(0, 10, 1000)
x = x[1:]
y = list(map(f, x))
plt.plot(x, y)
plt.show()
```

```python
from scipy.interpolate import UnivariateSpline

def drawY(x):
    y = list(map(exp, x))
    p, e = np.histogram(y, b, density=True)
    e = e[:-1] + (e[1]-e[0])/2
    while e[-1] > 10:
        e = e[:-1]
        p = p[:-1]    
    plt.plot(e, p)
    plt.show()

n = 10000
b = 1000
x = np.random.default_rng().normal(0, 1, n)
drawY(x)
```

```python
def box_muller(u1, u2):
    r = -2*log(u1)
    theta = 2*np.pi*u2
    return sqrt(r)*cos(theta), sqrt(r)*sin(theta)

u = np.random.rand(n)
x = []
for i in range(0, n, 2):
    x.extend(box_muller(u[i], u[i+1]))
drawY(x)
```
