        تخمینگر بیشینه درست نمایی (Maximum Likelihood) با بیشینه کردن توزیع شرطی  $f_{X|\Theta}(x)$ پارامتر $\Theta$ را تخمین میزند، یعنی  $\hat{\theta}_{MLE} = \arg \max_{\theta} f_{X|\Theta = \theta}(x|\Theta = \theta) $. کدی بنویسید که تعدادی نقطه ($n$ نقطه) را در صفحه (نقاط 2 بعدی) را با توزیع یکنواخت بین 0 و 1 تولید کند. حال باید پارامترهای میانگین و ماتریس کواریانس برای $k$ توزیع نرمال تولید کنید. (مقادیر میانگین ها خود دارای توزیع یکنواخت بین 0 و 1 باشند و برای مقادیر ماتریس کواریانس به بندهای الف تا د دقت کنید.)  حال میخواهیم هر نقطه را به دقیقا یک توزیع نسبت دهیم. برای این کار از تخمینگر بیشینه درست نمایی (Maximum Likelihood) استفاده میکنیم. به این منظور برای هر نقطه باید $f_{(X|\mu, \Sigma)}(x|\mu, \Sigma)$ را برای هر $k$ توزیعی که پارامترهای آن را مشخص کردید محاسبه کنید (توجه کنید چون نقاط دو بعدی هستند $X = (X_0, X_1)$). برای اینکار میتوانید از  تابع چگالی احتمال مربوط به توزیع گاووسی دو متغیره استفاده کنید. سپس با تعیین کردن توزیعی که بیشترین $f_{(X|\mu, \Sigma)}(x|\mu, \Sigma)$ مربوط را برای یک نقطه دارد، آن نقطه را به آن توزیع نسبت دهید. نقاط را به صورت scatter plot در صفحه رسم کنید، به همه ی نقاطی که به یک توزیع نسبت داده شده اند یک رنگ منحصر به فرد نسبت دهید. همچنین با علامتی متفاوت نقاط میانگین توزیع های گاوسی را نیز مشخص کنید. به علاوه برای تمام نقاط توزیع های شرطی $f_{X|\mu, \Sigma} (x|\mu_i, \Sigma_i) \  i = 1,...,k$ را رسم کنید.

```python
from numpy.random import rand, randn
import numpy as np
from mpl_toolkits import mplot3d
from matplotlib import pyplot as plt
```

```python
def estimate_likelihood(X, mean, sigma):
    p = np.zeros((X.shape[0], 1), dtype = float)
    for i in range(X.shape[0]) : 
        p[i] = (1 / np.sqrt((np.pi * 2) ** np.linalg.det(sigma))) * np.exp(-0.5 * np.matmul(
        np.matmul(np.matrix.transpose(X[i] - mean) , np.linalg.inv(sigma)), (X[i] - mean)))
    return p

def plot_max_likelihood(X, means, sigmas, colors, info):
    print(sigmas)
    P = np.hstack([estimate_likelihood(X, m, s) for m, s in zip(means, sigmas)])
    classes = P.argmax(1)
    
    for i, color in enumerate(colors):
        plt.scatter(X[classes == i][:, 0], X[classes == i][:, 1], c=color)
    plt.scatter(means[:, 0], means[:, 1], s=120, c=colors, marker='*', edgecolors='k')
    
    plt.xlabel('X0')
    plt.ylabel('X1')
    plt.title(f'n = {X.shape[0]}, k = {len(colors)}, {info}')
    plt.gca().set_aspect('equal', 'box')
    plt.show()
    
    fig = plt.figure(figsize=(20, 10))
    for i, color in enumerate(colors):
        ax = fig.add_subplot(2, 3, i+1, projection='3d')
        ax.scatter3D(X[:,0].ravel(), X[:,1].ravel(), P[:,i].ravel(), c=color);
        ax.set_xlabel('X_0')
        ax.set_ylabel('X_1')
        ax.set_zlabel('Conditional PDF');
        ax.set_title('$f(X_0,X_1|\Sigma_' + str(i+1) +', \mu_' +str(i+1) + ')$', fontsize = 25);
    plt.tight_layout()
```

```python
n, d, k = 4000, 2, 4
X = (rand(n, d)) # test points 

means = (rand(k, d))
colors = ['red', 'blue', 'green', 'yellow']
```

```python
A = (rand(k, d, d) - 0.5)/2

sigmas = A @ A.transpose(0, 2, 1)
plot_max_likelihood(X, means, sigmas, colors, 'different random sigmas')
```

```python
A = (rand(1, d, d) - 0.5)/2
S = A @ A.transpose(0, 2, 1)
sigmas = np.concatenate([S] * k)
plot_max_likelihood(X, means, sigmas, colors, 'same random sigmas')
```

```python
S = np.expand_dims(np.diag([0.1, 10]), axis=0)
sigmas = np.concatenate([S] * k)

plot_max_likelihood(X, means, sigmas, colors, 'same diagonal sigmas')
```

```python
S = np.expand_dims(np.identity(d), axis=0)
sigmas = np.concatenate([S] * k)

plot_max_likelihood(X, means, sigmas, colors, 'identity sigmas')
```
